Markov Chains

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
P = np.array([[0, 1/3, 1/2, 1/6],
              [1/3, 0, 1/2, 1/6],
              [1/6, 1/6, 0, 2/3],
              [1/3, 1/3, 1/3, 0]])
v = np.array([1/4, 1/2, 1/8, 1/8])
N = 10
traj = np.zeros(N)

traj[0] = np.random.choice(range(P.shape[1]), p=v)     # initial state
for i in range(1, N):
  traj[i] = np.random.choice(range(P.shape[1]), p=P[traj[i - 1].astype(int)])

plt.plot(traj, ':d')
plt.title('A realization of Markov Chain evolving in time')
plt.xlabel('Time')
plt.ylabel('States')
plt.show()
```
